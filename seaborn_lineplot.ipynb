```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0)
df.head()
```

```python
sns.lineplot(data=df)
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.title('amerika muzeylariga tashriflar')
plt.xlabel('sana')
plt.ylabel('odam soni')
plt.xticks(rotation=90) # x o'qidagi yozuvlarni 90 gradusga aylantiramiz
plt.grid()
sns.lineplot(data=df)
plt.show()
```
