```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
!git clone https://github.com/anvarnarz/praktikum_datasets.git
```

```python
df = pd.read_csv('/content/praktikum_datasets/usa_cars.csv', index_col=0)
df.head()
```

```python
df.drop(columns=['model','title_status','color','vin','lot','country','condition'],inplace=True)
df.head()
```

```python
df2 = pd.pivot_table(df, index = 'price')
df2
```

```python
df2.drop(columns='year', inplace=True)
df2
```

```python
plt.figure(figsize=(20,6))
plt.grid()
sns.lineplot(data=df)
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.grid()
sns.lineplot(data=df2)
plt.show()
```

```python
df3 = df2.apply(lambda x: x/1000)
df3
```

```python
plt.figure(figsize=(20,6))
plt.grid()
plt.xticks(rotation=90)
sns.barplot(x=df3['mileage'], y=df3.index)
plt.show()
```
