```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/drive/MyDrive/praktikum_datasets/merc.csv')
df.head()
```

```python
df = df[df.year>2000]
```

```python
plt.figure(figsize=(16,6))
sns.scatterplot(data=df, x='mileage', y='price')
#sns.regplot(data=df, x='year', y='price')  #osuvchi yoki negativ korilliyatsiyani korsatadi
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(data=df, x='mileage', y='price', hue='transmission',s=100)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(data=df, x='mileage', y='price', hue='transmission',style='fuelType',s=100)
plt.show()
```
