```python
x  = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
```

```python
plt.plot(x,y, linewidth=2.0, linestyle='--')
plt.show()
```

```python
plt.plot(x, y,
 linewidth=1.5,
 linestyle='-.',
 color='#8E44AD',
 alpha=0.7,
 marker='^')
plt.show()
```

```python
plt.plot(x, y,
 linewidth=1.5,
 linestyle='-.',
 color='#8E44AD',
 alpha=0.7,
 marker='h')
plt.title('sin(x) grafigi')
plt.xlabel('pi')
plt.ylabel('sin(x)')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(x, y,
 linewidth=1.5,
 linestyle='-.',
 color='#8E44AD',
 alpha=0.7,
 marker='h')
plt.title('sin(x) grafigi')
plt.xlabel('pi')
plt.ylabel('sin(x)')
plt.grid()
plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)
```

```python
plt.figure(figsize=(12,6))
plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')


plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')

plt.title('sin(x) va cos(x) grafiklari')
plt.xlabel('pi')
plt.legend(['sin(x)','cos(x)'], loc=1, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--') # setkani o'zgartirish
plt.show()

```

```python
x = np.arange(0, 4*np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize=(12,6))
plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')


plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')

plt.title('sin(x) va cos(x) grafiklari')
plt.xlabel('pi')
plt.legend(['sin(x)','cos(x)'], loc=1, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--') # setkani o'zgartirish
plt.show()


plt.savefig('sin-cos-grafik.png') # rasmni saqlab qoyamiz
```
